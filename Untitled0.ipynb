```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the dataset
df = pd.read_csv('parkinsons_updrs.data')
df = df.drop(columns=['subject#', 'motor_UPDRS'])
df
```

```python
import pandas as pd
import pycaret
pycaret.__version__
```

```python
from pycaret.regression import *
setup(data=df, target='total_UPDRS', session_id=123, preprocess=True)
```

```python
best = compare_models()
```

```python
best_model = create_model('catboost')
```

```python
plot_model(best_model, plot='residuals')  # Residuals plot
plot_model(best_model, plot='error')  # Error plot
plot_model(best_model, plot='feature')  # Feature importance
```

```python
evaluate_model(best_model)
```

```python
predictions = predict_model(best_model)

predictions.head()
```
